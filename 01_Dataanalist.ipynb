```python
#dtypes 속성이나 info()함수를 이용해서 데이터 프레임을 구성하는 데이터 타입을 확인하고
# 필요하면 타입 변경한다
pre_sale.dtypes
```

```python
#연도와 월은 연산에 사용되는 데이터가 아니라 구분에 사용하는 데이터 이므로 문자열 형태로 변환한다
pre_sale['연도']=pre_sale['연도'].astype(str)
pre_sale['월']=pre_sale['월'].astype(str)
```

```python
#분양가격 데이터 타입을 숫자로 변경하고 평당 분양가격을 계산한다
#to_numeric() errors 속성을 coerce로 지정하면 오류 발생 데이터를 누락값으로 변경
pre_sale['분양가격(㎡)']=pd.to_numeric(pre_sale['분양가격(㎡)'],errors='coerce' )
pre_sale['평당 분양가격']=pre_sale['분양가격(㎡)']*3.3
pre_sale
```

전국 지역별 평균 분양가격

```python
#규모 구분이 전체로 되어 있는 금액으로만 연도별 금액을 살펴보자
region_year_all=pre_sale[pre_sale['규모구분']=='전체']
region_year_all
```

```python
#피벗테이블에 결과는 두번쨰 인덱스가 피벗테이블에 인덱스로 사용도ㅣ므로 피벗테이블
#인덱스 데이터를 사용하기 위해서 인덱스를 다시 지정하려면 reset_index()사용
#region_year=region_year_all.pivot_table('평당 분양가격','지역명','연도')
region_year=region_year_all.pivot_table('평당 분양가격','지역명','연도').reset_index()
region_year
```

전국 지역별 평균 분양가격 변동 금액

```python
region_year['변동액']=region_year['2018']-region_year['2015']
region_year
```

```python
max_delta_price=region_year['변동액'].max()
print('상승액이 가장 큰 지역은 제주도로 상승액은 평당 {:,.1f}이다.'.format(max_delta_price))
```

```python
min_delta_price=np.min(region_year['변동액'])
min_delta_price
```

```python
mean_delta_price=region_year['변동액'].mean()
mean_delta_price
```

```python
#전국 지역별 평균 분양가격 시각화
#ggplot(데이터프레임,aes(x='x축 데이터',y='y축 데이터))
ggplot(region_year_all,aes(x='지역명',y='평당 분양가격',fill='연도'))\
   +geom_bar(stat='identity',position='dodge')\
   +ggtitle('2015~2018년 신규 민간 아파트 분양 가격')\
   +theme(text=element_text(family=font_name), figure_size=(10, 6))
```
